```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
import matplotlib.pyplot as plt
```

```python
wcss=[]
for i in range(2,11):
    kmean=KMeans(n_clusters=i,random_state=42)
    kmean.fit(df_final)
    wcss.append(kmean.inertia_)
```

```python
plt.plot(range(2, 11), wcss)
plt.title('Elbow Plot')
plt.xlabel('k_value')
plt.ylabel('wcss')
```

```python
knee=KneeLocator(range(2,11),wcss,curve="convex",direction="decreasing")
```

```python
knee.knee
```

```python
kmeans=KMeans(n_clusters=4,random_state=42)
```

```python
df_final['clusters']=kmeans.fit_predict(df_final)
```

```python
df_final
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

```python
for i in df_final['clusters'].unique():
    data=df_final[df_final['clusters']==i]
    features=data.drop(columns=["price","clusters"])
    label=data["price"]
    scaler=StandardScaler()
    scaled_features=pd.DataFrame(scaler.fit_transform(features),columns=features.columns)
    x_train,x_test,y_train,y_test=train_test_split(scaled_features,label,test_size=0.2,random_state=42)
    rf=RandomForestRegressor()
    rf.fit(x_train,y_train)
    pred_rf=rf.predict(x_test)
    print(f'cluster-{i} {r2_score(y_test,pred_rf)}\n')
    
    
```
